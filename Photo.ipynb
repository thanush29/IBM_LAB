```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from google.colab.patches import cv2_imshow

# Read the color image
pic = cv2.imread('/content/download (1).jpeg')
plt.imshow(cv2.cvtColor(pic, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()

# Read the image in grayscale
pic = cv2.imread('/content/download.jpeg', cv2.IMREAD_GRAYSCALE)
plt.imshow(pic, cmap='gray')
plt.title('Grayscale Image')
plt.show()

# Flip the image upside down
flip_pic = np.flipud(pic)
plt.imshow(flip_pic, cmap='gray')
plt.title('Flipped Image')
plt.show()

# Rotate the image by 45 degrees
rot_pic = ndimage.rotate(pic, 45)
plt.imshow(rot_pic, cmap='gray')
plt.title('Rotated Image (45 degrees)')
plt.show()

# Apply Gaussian filter with sigma=5
gauss_pic = ndimage.gaussian_filter(pic, 5)
plt.imshow(gauss_pic, cmap='gray')
plt.title('Gaussian Filtered Image')
plt.show()

# Create the negative of the image
pic2 = 255 - pic
cv2_imshow(pic2)  # Use cv2_imshow instead of cv2.imshow
print("Negative Image Displayed")

# Image properties
print(f"Image Shape: {pic.shape[0:2]}")
pixels = pic[100, 100] if pic.shape[0] > 100 and pic.shape[1] > 100 else "Pixel out of bounds"
print(f"Pixel Value at (100, 100): {pixels}")
print(f"Image Shape: {pic.shape}")
print(f"Total Pixels: {pic.size}")
print(f"Mean Pixel Value: {pic.mean()}")
print(f"Max Pixel Value: {pic.max()}")
print(f"Min Pixel Value: {pic.min()}")
```
